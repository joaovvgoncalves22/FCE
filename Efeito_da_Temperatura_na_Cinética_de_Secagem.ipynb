```python
from dataclasses import dataclass
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parâmetros

@dataclass
class ParametrosSecagem:

    # Umidade de equilíbrio
    Xe: float = 0.05

    # Temperatura de secagem
    T: float = 60.0
```

```python
# Função


def calcula_k(T):
    """
    k(T) = 0.2 * [1 + 0.02*(T - 60)]
    """

    return 0.2 * (1 + 0.02 * (T - 60))
```

```python
# Modelo diferencial


def modelo_secagem(t, X, params):

    # Calcula k
    k = calcula_k(params.T)

    # Equação diferencial
    dXdt = -k * (X - params.Xe)

    return dXdt
```

```python
# Função de simulação

def simular_secagem(tempo, X0, params):

    solucao = solve_ivp(
        modelo_secagem,
        t_span=(tempo[0], tempo[-1]),
        y0=[X0],
        t_eval=tempo,
        args=(params,),
        method='RK45'
    )

    return solucao.t, solucao.y[0]
```
