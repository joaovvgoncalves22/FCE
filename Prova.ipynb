```python
from dataclasses import dataclass
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt
```

```python
@dataclass
class ParametrosBioreator:
  #parametros do modelo
  V:float = 5 #mgL-1h-1
  K:float = 20 #mgL-1
  E:float = 30000 #jmol-1
  R:float = 8.314 #jmol-1K-1
  T:float = 298.15 #k
  S0 = 100 # mg/L
  t0= 30 # horas
```

```python
def Bioreator_modelo(t, S, params,T):
    Vmax = Vmax_Arrhenius(T, params)
    dSdt = -Vmax * S / (params.K + S)
    return dSdt
```

```python
def Vmax_Arrhenius(T_kelvin, params):

  Vmax = params.V * np.exp(-params.E / (params.R * T))
  return Vmax
```

```python
def Simula_Bioreator(t0=30, S0=100, T=298.15, params=None):
    if params is None:
        params = ParametrosBioreator()

    solv = solve_ivp(
        fun=Bioreator_modelo,
        t_span=(t[0], t[-1]),
        y0=[S0],
        args=(params, T)
    )

    return solv.t, solv.y[0]
```
